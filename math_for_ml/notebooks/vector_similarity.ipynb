```
import numpy as np

vectors = {
    "hello": np.array([1, 0, 1]),
    "bye": np.array([0, 1, 0]),
    "ok": np.array([1, 1, 1]),
    "world": np.array([0, 0, 1]),
    "python": np.array([1, 1, 0]),
    "code": np.array([0, 1, 1])
}

def cosine_similarity(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

def similarity(word1, word2):
    return cosine_similarity(vectors[word1], vectors[word2])

print("Similarity hello-bye:", similarity("hello", "bye"))
print("Similarity hello-ok:", similarity("hello", "ok"))
print("Similarity python-code:", similarity("python", "code"))
print("Similarity hello-world:", similarity("hello", "world"))

hello_world = vectors["hello"] + vectors["world"]
print("Similarity hello vs 'hello world':", cosine_similarity(vectors["hello"], hello_world))

```
